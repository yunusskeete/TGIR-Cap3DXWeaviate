```python
def process_image_objects(
    object_folder: Path,
    descriptions_dict: Dict[str, str],
    collection: Collection,
    existing_uuids: List[str],
    file_extension: str = ".png",
    image_filter: Optional[str] = None,
) -> List[str]:
    """
    Processes the images in the given folder, creates DataObject instances, and inserts them into the Weaviate collection.

    Args:
        object_folder (Path): The folder containing image files to process.
        descriptions_dict (Dict[str, str]): A dictionary mapping folder names to object descriptions.
        collection (Collection): The Weaviate collection to insert the DataObjects into.
        existing_uuids (List[str]): A list to append new UUIDs for the created DataObjects.
        file_extension (str): The file extension for image files. Default is ".png".
        image_filter (Optional[str]): An optional filter for image files (e.g., to exclude certain files).

    Returns:
        List[str]: A list of UUIDs for the inserted DataObjects.
    """
    # Description for the object
    object_description = descriptions_dict.get(object_folder.name, "")

    image_objects_buffer = []
    for object_image_file in object_folder.iterdir():
        if object_image_file.suffix == file_extension and (
            image_filter is None or image_filter not in object_image_file.name
        ):
            # Convert the image to base64
            with object_image_file.open("rb") as file:
                image_b64 = base64.b64encode(file.read()).decode("utf-8")

            # Build the image object properties
            image_obj = {
                "image": image_b64,
                "description": object_description,
                "datasetUID": f"{object_image_file.parent.name}_{object_image_file.name}",
            }

            image_object_uuid = generate_uuid5(image_obj["datasetUID"])

            existing_uuids.append(image_object_uuid)

            # Create DataObject and add to buffer
            image_objects_buffer.append(
                DataObject(
                    properties=image_obj,
                    uuid=image_object_uuid,
                )
            )

    # Insert the buffer into the collection
    try:
        batch_return = collection.data.insert_many(image_objects_buffer)
    except weaviate.exceptions.WeaviateInsertManyAllFailedError as e:
        print(f"Weaviate InsertManyAllFailedError: {e}")
        return []

    # Check for errors during insertion
    if batch_return.has_errors:
        print(f"Failed to insert some objects")
        for idx, error in batch_return.errors.items():
            print(f"Error at index {idx}: {error.message}")

    return existing_uuids
```
